```python

from llama_index import Prompt
# Define a custom prompt
template = (
    "We have provided context information below. \n"
    "---------------------\n"
    "{context_str}"
    "\n---------------------\n"
    "Given this information, please answer the question and each answer should start with code word Doc chat:. And if the answer is not in given context should reply with sorry. {query_str}\n"
)
qa_template = Prompt(template)
```

```python
# Create a query engine from the index
query_engine = index.as_query_engine(text_qa_template=qa_template, response_mode='tree_summarize', similarity_top_k=5)
```

```python
import os
import re
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
from nltk.corpus import wordnet
import concurrent.futures

# Cache for storing previous query results
query_cache = {}

def preprocess_query(query):
    # Tokenize, remove non-alphanumeric characters, and apply stemming
    tokens = word_tokenize(query)
    stemmer = PorterStemmer()
    tokens = [stemmer.stem(token) for token in tokens if token.isalnum()]
    return ' '.join(tokens)

def expand_query(query):
    # Expand query with synonyms from WordNet
    synonyms = set()
    for syn in wordnet.synsets(query):
        for lemma in syn.lemmas():
            synonyms.add(lemma.name())
    return ' '.join(synonyms)
```

```python
def answer_question(query):
  try:
      # Check if the result is already in the cache
      preprocessed_query = preprocess_query(query)
      if preprocessed_query in query_cache:
          return query_cache[preprocessed_query]

      # Expand the query with synonyms
      expanded_query = expand_query(preprocessed_query)
      response = query_engine.query(preprocessed_query)

      # Cache the result
      query_cache[preprocessed_query] = response

      return response
  except Exception as e:
    print(f"Error answering question '{query}': {e}")
    return "Sorry, I encountered an error while processing your request."
```

```python
# Test the chatbot with different queries
queries = [
    "What are the types of machine learning?",
    "Explain machine learning categories.",
    "Tell me about different machine learning techniques.",
    "Types of ML algorithms?"
    # "provide examples of common techniques used in preparing data for analysis"
]

for q in queries:
    print(f"Query: {q}")
    print("Answer:", answer_question(q))
    print("="*50)
```

```python
import os
import fitz  # PyMuPDF

def extract_text_from_pdf(pdf_path):
    doc = fitz.open(pdf_path)
    text = ""
    for page_number in range(doc.page_count):
        page = doc[page_number]
        text += page.get_text()
    return text

def split_sections(text):
    subject_topics = {
        'mathematics': ['math', 'algebra', 'calculus'],
        'ai': ['machine learning', 'supervised', 'machine', 'learning'],
        'biology': ['biology', 'genetics', 'ecology'],
        'physics': ['quantum', 'mechanics', 'relativity']
    }

    sections = {subject: [] for subject in subject_topics}

    current_section = 'general'  # Default section for text not matched to any subject

    for line in text.split('\n'):
        line_lower = line.lower()

        for subject, keywords in subject_topics.items():
            if any(keyword in line_lower for keyword in keywords):
                current_section = subject
                break

        sections[current_section].append(line)

    return sections

def save_sections(sections, output_folder):
    for section_name, section_content in sections.items():
        filename = f"{section_name}_section.txt"
        filepath = os.path.join(output_folder, filename)
        with open(filepath, 'w', encoding='utf-8') as file:
            file.write('\n'.join(section_content))

if __name__ == "__main__":
    pdf_path = "/content/data/Case_Study_1_Refined_v2.pdf"
    output_folder = "/content/output"

    # Step 1: Extract text from PDF
    pdf_text = extract_text_from_pdf(pdf_path)

    # Step 2 & 3: Split the content into sections based on subject topics
    sections = split_sections(pdf_text)

    # Display and save sections (you can remove this if you don't need it)
    for subject, content in sections.items():
        print(f"Subject: {subject}")
        print("\n".join(content))  # Display the content for each subject
        print("=" * 50)

    # Step 4: Save each section
    save_sections(sections, output_folder)
```
